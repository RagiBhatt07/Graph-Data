Note:

    The random network model, as formulated by Gilbert, has two parameters: the number of nodes, and the link probability. As in the book text, the algorithm for creating this network is as follows:

    1. Select a pair of nodes, say i and j.
    2. Generate a random number r between 0 and 1. If r < p, then add a link between i and j.
    3. Repeat (1) and (2) for all pairs of nodes.

```python
def random_graph(N, p): #N is the number of nodes and p is the probability of an edge
    G = nx.Graph()
    G.add_nodes_from(range(N))

    for i, j in itertools.combinations(G.nodes, 2):
        r = random.random()
        if r < p:
            G.add_edge(i, j)

    return G
```

```python
G_random = random_graph(34, 0.1390)
nx.draw(G_random, with_labels=True, node_color='pink', edge_color='yellow')
plt.title('Erdős–Rényi Random Graph with N nodes and probability p')

print("Number of nodes:", len(G_random.nodes()))
print("Number of edges:", len(G_random.edges()))
```

### Section 3 Counting Triangles in Graphs

An important measure in graph analysis is the number of triangles it contains. A triangle of a graph
G is a connected subgraph of G having 3 nodes, or a 3-clique.
Your task is to count the number of triangles in the graph, without using the NetworkX implementation. Proceed in two steps:


```python
#Counting the traingles to store them in the dictionary with the node as the key and the number of triangles as the value
triangle_count_custom = {node: 0 for node in G.nodes()}

for node in G.nodes():
    neighbors = list(G.neighbors(node))
    # Iterate through pairs of neighbors
    for i in range(len(neighbors)):
        for j in range(i + 1, len(neighbors)):
            # Check if the neighbors are connected, forming a triangle with the original node
            if G.has_edge(neighbors[i], neighbors[j]):
                # Increment the triangle count for each vertex of the triangle
                triangle_count_custom[node] += 1
                
triangle_count_custom = {node: count for node, count in triangle_count_custom.items()}
triangle_count_custom
```

#### For testing, compare your output with that of the networkx.triangles method.

```python
nx_triangle_count = nx.triangles(G)

print("NetworkX triangle counts: ", nx_triangle_count)
print("Custom implementation triangle counts: ", triangle_count_custom)

# Check if the counts are equal
comparison = all(nx_triangle_count[node] == triangle_count_custom[node] for node in G.nodes())
print("Do the results match?: ", comparison)
```

#### Using the dictionary from step 1, output the total number of triangles in the graph.

### Section: 4 Community Detection
    Implement the divisive community detection (Ravasz) as presented in the course slides. In short, the
    following steps should be followed, starting from a single community, i.e., the connected graph:


```python
import networkx as nx

def girvan_newman(G_original):
    G = G_original.copy()
    best_modularity = -1
    best_communities = None

    while len(G.edges) > 0:
        # 1. Compute the betweenness centrality of all edges
        edge_betweenness = nx.edge_betweenness_centrality(G)
        
        # 2. Remove the edge with highest centrality
        edge_to_remove = max(edge_betweenness, key=edge_betweenness.get)
        G.remove_edge(*edge_to_remove)

        # 3. Find communities (connected components)
        communities = list(nx.connected_components(G))

        # 4. Compute modularity
        modularity = nx.algorithms.community.modularity(G_original, communities)
        
        # Track the partition with the highest modularity
        if modularity > best_modularity:
            best_modularity = modularity
            best_communities = communities

    return best_communities, best_modularity

# Assuming G is your graph
best_partition, max_modularity = girvan_newman(G)

print(f"Partition with maximal modularity: {best_partition}")
print(f"Maximal Modularity: {max_modularity}")
```

```python
# Assuming G is your graph and girvan_newman() is the function implemented previously
best_partition, max_modularity = girvan_newman(G)

# Create a mapping of node to community number
community_map = {}
for i, community in enumerate(best_partition):
    for node in community:
        community_map[node] = i

# Draw the original graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)  # positions for all nodes

# For each community, draw the nodes in that community with a specific color
for i, community in enumerate(best_partition):
    nx.draw_networkx_nodes(G, pos, nodelist=community, node_color=plt.cm.jet(i / len(best_partition)))

# Draw the edges and labels
nx.draw_networkx_edges(G, pos, alpha=0.5)
nx.draw_networkx_labels(G, pos)

plt.title("Communities in the Graph Corresponding to Maximum Modularity")
plt.show()
```
